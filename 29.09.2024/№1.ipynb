```python
import matplotlib.pyplot as plt
import numpy as np
x_measured = [0.358,0.507,0.685,1.003,1.444,1.651]
y_measured = [0.146,0.580,2.533,5.756,9.629,11.272]
plt.figure(figsize=(8,5), dpi=100)
x = [0.358, 1.651]
y = np.interp(x, x_measured, y_measured)
plt.scatter(x_measured, y_measured, marker='x')
plt.errorbar(x_measured, y_measured, yerr=0.2, xerr = 0.1, color = 'k', linestyle = 'None')
plt.plot(x,y, 'r')

plt.grid()

plt.show()
```
